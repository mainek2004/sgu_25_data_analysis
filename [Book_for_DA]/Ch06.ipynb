```python

import pandas as pd
import numpy as np
import os

os.makedirs("examples", exist_ok=True)

N = 10000
df_ex6 = pd.DataFrame({
    "key": np.random.choice(["one", "two", "three", "four"], size=N),
    "value": np.random.randn(N)
})
df_ex6.to_csv("examples/ex6.csv", index=False)

# Đọc lại theo chunksize
chunker = pd.read_csv("examples/ex6.csv", chunksize=1000)

tot = pd.Series([], dtype='int64')
for piece in chunker:
    tot = tot.add(piece["key"].value_counts(), fill_value=0)

tot = tot.sort_values(ascending=False)
print(tot)
```

```python
import os
from lxml import objectify
import pandas as pd

def make_mta_xml(path="datasets/mta_perf/Performance_MNR.xml"):
    # Đảm bảo thư mục tồn tại
    os.makedirs(os.path.dirname(path), exist_ok=True)

    # Dữ liệu mẫu
    rows = [
        {"Name": "Hudson",    "OnTime": 95.2, "Ridership": 120000},
        {"Name": "Harlem",    "OnTime": 93.5, "Ridership":  98000},
        {"Name": "New Haven", "OnTime": 91.8, "Ridership": 134000},
    ]

    # Tạo nội dung XML
    xml_content = """\n\n"""
    for r in rows:
        xml_content += f"""  \n"""
    xml_content += ""

    # Ghi file
    with open(path, "w", encoding="utf-8") as f:
        f.write(xml_content)

    # Đọc lại
    parsed = objectify.parse(open(path, "rb"))
    root = parsed.getroot()

    # Chuyển sang DataFrame
    data = []
    for line in root.Line:
        data.append({
            "Name": str(line.Name),
            "OnTime": float(line.OnTime),
            "Ridership": int(line.Ridership)
        })
    df = pd.DataFrame(data)

    return df, path, root

# Gọi hàm
df_mta, xml_path, xml_root = make_mta_xml()
print("File đã tạo:", xml_path)
print(df_mta)
```

##Interacting with Databases

```python
import sqlite3

query = """
CREATE TABLE test
(a VARCHAR(20), b VARCHAR(20),
 c REAL,        d INTEGER
);"""

con = sqlite3.connect("mydata.sqlite")
con.execute(query)
con.commit()


```

```python
data = [("Atlanta", "Georgia", 1.25, 6),
        ("Tallahassee", "Florida", 2.6, 3),
        ("Sacramento", "California", 1.7, 5)]
stmt = "INSERT INTO test VALUES(?, ?, ?, ?)"

con.executemany(stmt, data)
con.commit()

```

```python
cursor = con.execute("SELECT * FROM test")
rows = cursor.fetchall()
rows
```

```python
cursor.description
pd.DataFrame(rows, columns=[x[0] for x in cursor.description])


```

```python
import sqlalchemy as sqla
db = sqla.create_engine("sqlite:///mydata.sqlite")
pd.read_sql("SELECT * FROM test", db)
```
