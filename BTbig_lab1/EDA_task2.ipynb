# **Xử lý và trực quan hóa dữ liệu**

## Ôn tập lý thuyết

###Trực quan hóa dữ liệu có vai trò gì trong phân tích dữ liệu?

  Biểu diễn trực quan: Biến dữ liệu thô (số liệu, bảng biểu khó nhìn) thành hình ảnh (biểu đồ, đồ thị) giúp người phân tích dễ nhận ra xu hướng, quy luật và mối quan hệ.

  Phát hiện mẫu (patterns) và xu hướng (trends): Qua biểu đồ (line, bar, scatter, histogram, boxplot...), ta có thể thấy rõ dữ liệu tăng giảm theo thời gian, sự khác biệt giữa các nhóm, hay phân bố của một biến.

  Phát hiện bất thường (outliers) và nhiễu: Những điểm dữ liệu "khác thường" thường rất khó nhận ra khi nhìn bảng số, nhưng trên đồ thị lại nổi bật ngay.

  Hỗ trợ so sánh: Dễ dàng so sánh giữa các nhóm, danh mục hoặc các khoảng giá trị để rút ra kết luận.

  Tăng hiệu quả truyền đạt: Không chỉ giúp nhà phân tích hiểu nhanh hơn, mà còn giúp thuyết phục người khác (quản lý, khách hàng, đối tác) vì hình ảnh trực quan dễ tiếp nhận hơn số liệu khô khan.

### Tại sao trực quan hóa quan trọng trong khám phá dữ liệu (EDA)?

EDA là giai đoạn khám phá, hiểu rõ đặc điểm dữ liệu trước khi xây dựng mô hình. Trực quan hóa giúp:

  Hiểu phân bố dữ liệu: Ví dụ histogram hoặc boxplot giúp xem dữ liệu có lệch (skewed), có phân phối chuẩn hay không.

  Xem mối quan hệ giữa các biến: Scatter plot, heatmap cho thấy sự tương quan (correlation) hoặc mối liên hệ tuyến tính/phi tuyến.

  Phát hiện vấn đề dữ liệu: Thiếu dữ liệu (missing values), dữ liệu ngoại lai, sai lệch – những vấn đề này cần xử lý trước khi phân tích sâu.

  Định hướng phân tích tiếp theo: Sau khi nhìn thấy bức tranh dữ liệu, nhà phân tích sẽ biết nên chọn kỹ thuật thống kê, mô hình hoặc phương pháp tiền xử lý nào phù hợp.

###Các loại biểu đồ phổ biến (như histogram, scatter plot, boxplot, bar chart) được sử dụng trong các trường hợp nào?

  1. Histogram (biểu đồ tần suất)

Dùng khi: Muốn xem phân bố của một biến liên tục (continuous variable).

Ý nghĩa: Cho biết dữ liệu có phân bố chuẩn (normal) không, có lệch trái/phải (skewed) không, có nhiều đỉnh (multimodal) không.

Ví dụ: Phân bố điểm thi của sinh viên, độ tuổi khách hàng, mức thu nhập.

  2. Scatter plot (biểu đồ phân tán)

Dùng khi: Muốn xem mối quan hệ giữa hai biến liên tục.

Ý nghĩa: Thể hiện sự tương quan (tích cực, tiêu cực, hay không tương quan), phát hiện outliers hoặc mô hình phi tuyến.

Ví dụ: Mối quan hệ giữa số giờ học và điểm thi, giữa chiều cao và cân nặng, giữa giá nhà và diện tích.

  3. Boxplot (biểu đồ hộp)

Dùng khi: So sánh phân bố của một biến giữa nhiều nhóm (categorical groups).

Ý nghĩa: Thể hiện trung vị, tứ phân vị (Q1, Q3), độ phân tán, outliers.

Ví dụ: So sánh lương của nhân viên theo giới tính, so sánh điểm thi giữa các lớp, so sánh thời gian giao hàng giữa các khu vực.

  4. Bar chart (biểu đồ cột)

Dùng khi: Biểu diễn dữ liệu phân loại (categorical).

Ý nghĩa: So sánh tần suất hoặc giá trị trung bình giữa các nhóm.

Ví dụ: Số lượng khách hàng theo từng thành phố, doanh số theo từng sản phẩm, số người chọn mỗi loại dịch vụ.


###Làm thế nào để chọn loại biểu đồ phù hợp với đặc điểm của dữ liệu (ví dụ: dữ liệu phân loại, dữ liệu số, dữ liệu thời gian)?

1. Dữ liệu phân loại (Categorical Data)

Ví dụ: giới tính (Nam/Nữ), ngành học, loại sản phẩm.
Thường quan tâm đến tần suất hoặc tỷ lệ của từng nhóm.

Bar chart (cột) → so sánh số lượng hoặc giá trị trung bình giữa các nhóm.

Pie chart (tròn) → hiển thị tỷ lệ phần trăm (chỉ nên dùng khi số nhóm ít).

Count plot (bar chart đặc biệt) → đếm số lượng quan sát theo từng nhóm.


2. Dữ liệu số (Numerical / Continuous Data)

Ví dụ: điểm thi, thu nhập, chiều cao, cân nặng.
Thường quan tâm đến phân bố, xu hướng, mối quan hệ.

Histogram → xem phân bố (chuẩn, lệch trái/phải, nhiều đỉnh).

Boxplot → tóm tắt phân bố, phát hiện outlier, so sánh giữa các nhóm.

Scatter plot → tìm mối quan hệ giữa 2 biến số.

Line chart → khi muốn theo dõi xu hướng biến số liên tục.

3. Dữ liệu thời gian (Time Series Data)

Ví dụ: doanh số theo tháng, giá cổ phiếu theo ngày, nhiệt độ theo giờ.
Thường quan tâm đến xu hướng (trend), chu kỳ (seasonality), biến động (fluctuation).

Line chart → biểu diễn xu hướng qua thời gian.

Area chart → tương tự line nhưng nhấn mạnh phần diện tích, thường dùng cho dữ liệu tích lũy.

Candlestick chart → thường dùng trong tài chính (giá mở, cao, thấp, đóng cửa).

4. Dữ liệu kết hợp (Categorical + Numerical + Time)

Khi dữ liệu vừa có nhóm, vừa có số, vừa có yếu tố thời gian.

Boxplot theo nhóm → so sánh phân bố điểm thi theo giới tính.

Line chart nhiều series → so sánh doanh số nhiều sản phẩm theo thời gian.

Stacked bar chart → doanh thu từng danh mục sản phẩm theo năm.

###Những nguyên tắc thiết kế nào cần tuân thủ để tạo ra một biểu đồ trực quan hóa dễ hiểu và hiệu quả?

1. Xác định mục tiêu rõ ràng

Mỗi biểu đồ cần trả lời một câu hỏi cụ thể: xu hướng, phân bố, so sánh hay mối quan hệ.

Không vẽ quá nhiều thông tin trên cùng một biểu đồ, tránh làm người xem bị rối.

2. Chọn loại biểu đồ phù hợp

Histogram/Boxplot → phân bố dữ liệu.

Bar chart/Pie chart → so sánh nhóm phân loại.

Scatter plot → mối quan hệ giữa 2 biến số.

Line chart → xu hướng theo thời gian.
=> Chọn sai biểu đồ sẽ gây hiểu nhầm kết quả.

3. Giữ sự đơn giản và tối giản

Tránh quá nhiều màu sắc, hiệu ứng 3D, chi tiết thừa.

Dùng màu sắc có ý nghĩa (ví dụ: xanh dương = tích cực, đỏ = cảnh báo).

Giữ nền và lưới (grid) nhẹ nhàng, không lấn át dữ liệu.

4. Làm nổi bật thông tin quan trọng

Sử dụng màu sắc, độ dày đường, hoặc chú thích (annotation) để nhấn mạnh phần dữ liệu quan trọng.

Không nên làm tất cả đều nổi bật, chỉ làm nổi bật thông điệp chính.

5. Sắp xếp và chuẩn hóa dữ liệu

Nếu dùng bar chart → nên sắp xếp theo giá trị giảm dần để dễ so sánh.

Nếu dữ liệu có đơn vị khác nhau → chuẩn hóa hoặc dùng trục kép cẩn thận.

6. Thêm nhãn và tiêu đề rõ ràng

Luôn có: tiêu đề, nhãn trục (axis labels), đơn vị đo lường, chú thích (legend nếu cần).

Không để người xem phải đoán dữ liệu đại diện cho cái gì.

7. Tránh gây hiểu nhầm

Trục Y không nên “cắt bớt” giá trị gốc (0) trừ khi có lý do rõ ràng, vì dễ làm phóng đại sự khác biệt.

Tỷ lệ giữa các cột, vùng, hay vòng tròn phải chính xác với dữ liệu.

8. Nhất quán

Nếu nhiều biểu đồ trong cùng một báo cáo/dashboard → dùng cùng phong cách, bảng màu, font chữ, định dạng số liệu để người xem dễ theo dõi.

###Làm thế nào để tạo một biểu đồ đơn giản như histogram hoặc bar chart bằng Matplotlib? Bạn có thể chia sẻ đoạn code mẫu không?

Histogram (Biểu đồ tần suất)

```python
import matplotlib.pyplot as plt

# Dữ liệu điểm thi
scores = [5, 6, 7, 8, 5, 9, 6, 7, 8, 7, 6, 5, 9, 10, 8, 7, 6, 7, 8, 9]

# Vẽ histogram
plt.hist(scores, bins=5, color='skyblue', edgecolor='black')

# Thêm tiêu đề và nhãn
plt.title("Phân bố điểm thi")
plt.xlabel("Điểm")
plt.ylabel("Số lượng sinh viên")

plt.show()
```

Bar chart (Biểu đồ cột)

```python
import matplotlib.pyplot as plt

# Dữ liệu ngành học
majors = ['CNTT', 'Kinh tế', 'Ngôn ngữ', 'Quản trị']
students = [50, 40, 30, 20]

# Vẽ bar chart
plt.bar(majors, students, color='orange')

# Thêm tiêu đề và nhãn
plt.title("Số lượng sinh viên theo ngành học")
plt.xlabel("Ngành học")
plt.ylabel("Số lượng sinh viên")

plt.show()
```

### Chuẩn bị dữ liệu

```python
import pandas as pd

house_prices = pd.read_csv("/content/drive/MyDrive/Data_Analysis/BTbig_lab1/HousingPrices-Amsterdam-August-2021.csv")
house_prices = house_prices[['Zip', 'Price', 'Area', 'Room']]

# create col priceSqm
house_prices['PriceSqm'] = house_prices['Price'] / house_prices['Area']

# save new file .csv
house_prices.to_csv("/content/drive/MyDrive/Data_Analysis/BTbig_lab1/housePricesData.csv", index=False)
```

```python
pd.read_csv("/content/drive/MyDrive/Data_Analysis/BTbig_lab1/housePricesData.csv")
```

### Trực quan hóa dữ liệu với thư viện Matplotlib

```python
housePrices_sort = house_prices.sort_values(by='Price', ascending=False)
housePrices_sort.head()
```

Case 1: Basic

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
x = housePrices_sort['Zip'][0:10]
y = housePrices_sort['Price'][0:10]
plt.bar(x, y)
plt.show()
```

Case 2: Advanced 1


```python
plt.figure(figsize=(12,6))
plt.bar(x, y)
plt.title('Top 10 zip có giá nhà cao nhất', fontsize= 15 )
plt.xlabel('Zip code', fontsize = 12)
plt.xticks(fontsize = 10)
plt.ylabel('House prices in millions', fontsize = 12)
plt.yticks(fontsize = 10)
plt.show()
```

Case 3: Advanced 2

```python
fig, ax = plt.subplots(figsize=(40,18))
x = housePrices_sort['Zip'][0:10]
y = housePrices_sort['Price'][0:10]
y1 = housePrices_sort['PriceSqm'][0:10]
plt.subplot(1,2,1)
plt.bar(x,y)
plt.xticks(fontsize=17)
plt.ylabel('House prices in millions', fontsize=25)
plt.yticks(fontsize=20)
plt.title('Top 10 Areas with the highest house prices',
fontsize=25)
plt.subplot(1,2,2)
plt.bar(x,y1)
plt.xticks(fontsize=17)
plt.ylabel('House prices per sqm', fontsize=25)
plt.yticks(fontsize=20)
plt.title('Top 10 Areas with the highest house prices per sqm',
fontsize=25)
plt.show()
```

### Trực quan hóa dữ liệu với thư viện Seaborn

Case 1: Basic

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize= (12,6))
data = housePrices_sort[0:10]
sns.barplot(data= data, x= 'Zip',y = 'Price')
```

Case 2: Advanced 1

```python
plt.figure(figsize= (12,6))
data = housePrices_sort[0:10]
ax = sns.barplot(data= data, x= 'Zip',y = 'Price')
ax.set_xlabel('Zip code',fontsize = 15)
ax.set_ylabel('House prices in millions', fontsize = 15)
ax.set_title('Top 10 Areas with the highest house prices', fontsize= 20)
```

Case 3: view multiple perspectives at once


```python
fig, ax = plt.subplots(1, 2,figsize=(40,18))
data = housePrices_sort[0:10]
sns.set(font_scale = 3)
ax1 = sns.barplot(data= data, x= 'Zip',y = 'Price', ax = ax[0])
ax1.set_xlabel('Zip code')
ax1.set_ylabel('House prices in millions')
ax1.set_title('Top 10 Areas with the highest house prices')
ax2 = sns.barplot(data= data, x= 'Zip',y = 'PriceSqm',
ax=ax[1])
ax2.set_xlabel('Zip code')
ax2.set_ylabel('House prices per sqm')
ax2.set_title('Top 10 Areas with the highest price per sqm')
```
